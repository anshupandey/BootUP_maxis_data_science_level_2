```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [1.2,1.9,2.4,3.2,7.5,9.4,6.4,5.2,8.4,7.1,6.9,3.5,2.8]
price =[180,220,260,290,650,840,690,560,780,690,720,410,320]
x = np.array(area)
y = np.array(price)
```

<img src = "linear_regresion.png">

```python
m = 0.5
c = 0.8
LR = 0.002
for i in range(800):
    # feeding data and getting predictions
    yhat = m*x+c
    # calculating error
    cost = np.mean(np.square(y-yhat))
    # Gradient descent
    djdm = np.dot((yhat-y),x)*2
    djdc = np.dot((yhat-y),np.ones(len(x)))*2
    m = m - LR*djdm
    c = c - LR*djdc
    print(cost)
    
```

```python
print(m)
print(c)
```

```python
plt.scatter(x,y)
predictions = x*m + c
plt.plot(x,predictions,c='r')
plt.show()
```
