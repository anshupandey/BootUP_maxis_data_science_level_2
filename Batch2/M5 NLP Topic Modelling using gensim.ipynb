## Data cleaning

```python
# dropping complaints with not text or less than 20 characters
data = [comp for comp in data if len(str(comp))>20]
len(data)
```

```python
import nltk
import gensim
from gensim.utils import simple_preprocess
import gensim.corpora as corpora
from nltk.corpus import stopwords
```

```python
sw = stopwords.words("english")
simple_preprocess(data[0])
```

```python
# removing stopwords
#for i in range(len(data)):
 #   doc = []
  #  for w in simple_preprocess(data[i]):
   #     if w not in sw:
    #        doc.append(w)
    #data[i] = doc
```

```python
len(data)
```

```python
# Not removing stopwords
for i in range(len(data)):
    doc = []
    for w in simple_preprocess(str(data[i])):
        doc.append(w)
    data[i] = doc
```

```python
# make bigrams
bigrams = gensim.models.Phrases(data,min_count=5,threshold=100)
bigram_mod = gensim.models.phrases.Phraser(bigrams)
bigram_data = [bigram_mod[doc] for doc in data]
bigram_data[10]
```

```python
bigram_data[0]
```

```python
# lemmatization using spacy
import spacy
nlp = spacy.load("en", disable=['parser','ner'])

final_data = []
allowed_tags = ['NOUN','ADJ','VERB','ADV']

for doc in bigram_data:
    doc = nlp(" ".join(doc))
    final_data.append([token.lemma_ for token in doc if token.pos_ in allowed_tags])
    
```

```python
final_data[0]
```

```python
doc = nlp(" ".join(bigram_data[0]))
for token in doc:
    print(token,token.pos_)
```

## Topic Modelling

```python
dictionary = corpora.Dictionary(final_data)
corpus = [dictionary.doc2bow(doc) for doc in final_data]

model = gensim.models.ldamodel.LdaModel(corpus=corpus,id2word=dictionary,num_topics=10,
                                       random_state=5,update_every=1,chunksize=50,
                                       passes=10,alpha='auto',per_word_topics=True)
```

```python
len(dictionary.keys())
```

```python
for topic in model.print_topics():
    print(topic,"\n")
```
