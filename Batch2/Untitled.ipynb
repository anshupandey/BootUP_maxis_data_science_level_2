# Anomaly Detection

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pyod.utils import data
```

```python
# Generate some dummy data
xtr,ytr,xts,yts=data.generate_data(n_train=800,n_features=2,n_test=200,contamination=0.05,random_state=1054)
print(xtr.shape)
print(xts.shape)
print(ytr.shape)
print(yts.shape)
```

```python
plt.figure(figsize=(15,6))
plt.scatter(xtr[:,0],xtr[:,1],c=ytr)
plt.show()
```

## Anomaly Detection using Isolation Forest

```python
from sklearn.ensemble import IsolationForest
model = IsolationForest(n_estimators=50,max_features=2,random_state=5,contamination=0.1)
model.fit(xtr)
```

```python
ypred = model.predict(xts)
plt.figure(figsize=(15,6))
sns.scatterplot(x=xts[:,0],y=xts[:,1],hue=ypred)
plt.show()
```
