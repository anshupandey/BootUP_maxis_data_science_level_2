```python
auth = tweepy.OAuthHandler(api_key,api_secret)
auth.set_access_token(access_token,access_token_secret)

api = tweepy.API(auth)
```

```python
tweet_data = api.search(q="Maxis",count=50)
```

```python
clean_tweets = []
sentiments = []
for tw in tweet_data:
    tw = tw.text
    tw = re.sub("RT","",tw)
    tw = re.sub("@[a-zA-Z0-9_:]+","",tw)
    tw = re.sub("https://[a-zA-Z0-9._\/]+","",tw)
    tw = tw.strip()
    if len(tw)>20:
        print(tw)

        tw = TextBlob(tw)
        if tw.detect_language()!="en":
            print(tw.detect_language())
            tw = tw.translate(to="en")

            print(tw,"\n---------------------")
        
        clean_tweets.append(tw.raw)
        if tw.sentiment.polarity>0.1:
            print("Positive")
            sentiments.append("Positive")
        elif tw.sentiment.polarity<-0.1:
            print("Negative")
            sentiments.append("Negative")
        else:
            print("Neutral")
            sentiments.append("Neutral")
    
```

```python
plt.figure(figsize=(12,5))
sns.countplot(sentiments)
plt.show()
```

## Wordcloud Analysis

```python
tweettext = " ".join(clean_tweets)
```

```python
wd = wordcloud.WordCloud(width=1200,height=700,max_words=50).generate_from_text(tweettext)
plt.figure(figsize=(10,6))
plt.imshow(wd)
plt.show()
```
