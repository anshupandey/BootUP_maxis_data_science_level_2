# Linear Regression with numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [1.2,1.8,2.4,3.8,6.2,7.3,4.6,5.2,6.3,4.2,3.9,2.7,8.2,9.5,5.8]
price =[220,210,200,460,590,680,520,580,680,410,320,350,790,850,650]
```

```python
x = np.array(area)
y = np.array(price)
```

```python
m = 300
c = 10
LR = 0.002
for i in range(800):
    
    # feed data and get the predictions
    yhat = m*x + c
    
    # calculate error
    error = np.mean(np.square(y - yhat))
    
    # Apply Gradient descent to change values of m and c so as to minimize error
    dedm = 2 * np.dot((yhat-y),x)
    dedc = 2 * np.dot((yhat-y),np.ones(x.size))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(error)
    
    #plt.figure(figsize=(4,2))
    #plt.scatter(x,y)
    #plt.plot(x,yhat,c='r')
    #plt.show()
```

```python
print(m,c)
```

## Linear Regression using sklearn

```python
x = np.array(area).reshape(15,1)
y = np.array(price)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
print(model.coef_,model.intercept_)
```

```python
ypred = model.predict(x)
plt.figure(figsize=(8,4))
plt.scatter(x,y)
plt.plot(x,ypred,c='r')
plt.show()
```
